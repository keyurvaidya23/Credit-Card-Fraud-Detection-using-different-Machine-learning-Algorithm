##Splitting of data

```python
x=df.drop(columns='Class',axis=1)
y=df['Class']
```

```python
# Here, I'm using 70% of the data for training and 30% for testing.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=45)
```

##Model

##Random Forest

```python
rf=RandomForestClassifier()
model1=rf.fit(x_train,y_train)
y_pred=model1.predict(x_test)


print(f'classification report :{classification_report(y_test,y_pred)}')
print('*************************************')
print(f'confusion matrix :{confusion_matrix(y_test,y_pred)}')
print('*************************************')
print(f'accuracy score :{accuracy_score(y_test,y_pred)}')
```

##Logistic Regression

```python
lr=LogisticRegression(max_iter=1000)
model2=lr.fit(x_train,y_train)
y_pred2=model2.predict(x_test)


print(f'classification report :{classification_report(y_test,y_pred2)}')
print('*************************************')
print(f'confusion matrix :{confusion_matrix(y_test,y_pred2)}')
print('*************************************')
print(f'accuracy score :{accuracy_score(y_test,y_pred2)}')
```

##Decision Tree

```python
dt = DecisionTreeClassifier(random_state=45)

model_3 = dt.fit(x_train, y_train)

y_pred3 = model_3.predict(x_test)

print(f'Classification Report:\n{classification_report(y_test, y_pred3)}')
print('*************************************')

print(f'Confusion Matrix:\n{confusion_matrix(y_test, y_pred3)}')
print('*************************************')

print(f'Accuracy Score: {accuracy_score(y_test, y_pred3)}')
```

##Comparsion of algorithm

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming the predictions and true labels are stored in the variables as follows:
# y_pred (Random Forest), y_pred2 (Logistic Regression), y_pred3 (Decision Tree)

# Extracting metrics for each model
# Placeholder variables for the predictions - to be replaced with actual predictions
y_pred_rf = y_pred   # Random Forest predictions
y_pred_lr = y_pred2  # Logistic Regression predictions
y_pred_dt = y_pred3  # Decision Tree predictions

# Calculating metrics for Random Forest
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

# Calculating metrics for Logistic Regression
accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr)
recall_lr = recall_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr)

# Calculating metrics for Decision Tree
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)

# Preparing data for bar plot
models = ['Random Forest', 'Logistic Regression', 'Decision Tree']
accuracies = [accuracy_rf, accuracy_lr, accuracy_dt]
precisions = [precision_rf, precision_lr, precision_dt]
recalls = [recall_rf, recall_lr, recall_dt]
f1_scores = [f1_rf, f1_lr, f1_dt]

# Plotting
plt.figure(figsize=(12, 6))

# Bar width
barWidth = 0.15
# Position of bars on x-axis
r1 = np.arange(len(models))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

plt.bar(r1, accuracies, color='blue', width=barWidth, edgecolor='grey', label='Accuracy')
plt.bar(r2, precisions, color='red', width=barWidth, edgecolor='grey', label='Precision')
plt.bar(r3, recalls, color='green', width=barWidth, edgecolor='grey', label='Recall')
plt.bar(r4, f1_scores, color='orange', width=barWidth, edgecolor='grey', label='F1 Score')

plt.xlabel('Model', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(accuracies))], models)
plt.ylabel('Score')
plt.title('Comparison of Machine Learning Models')

plt.legend()
plt.show()

```
